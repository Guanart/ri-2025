```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de recall y precision
recall = [0.09, 0.09, 0.09, 0.18, 0.27, 0.27, 0.27, 0.27, 0.27, 0.36, 0.36, 0.36, 0.36, 0.45, 0.45]
precision = [1.00, 0.50, 0.33, 0.50, 0.60, 0.50, 0.43, 0.38, 0.33, 0.40, 0.36, 0.33, 0.31, 0.36, 0.33]
rank = [i for i in range(1, len(recall) + 1)]
```

```python
# Plot sin interpolar
# pagina 19 presentacion de baeza
plt.figure(figsize=(8, 5))
plt.plot(rank, recall, 'o-', label='Recall')
plt.plot(rank, precision, 'o-', label='Precision')
plt.xlabel('Rank')
plt.title('Gráfico sin interpolar')
plt.legend()
plt.grid(True)
plt.xticks(rank)  # Mostrar todos los valores de rank en el eje x
plt.yticks(np.arange(0, 1.1, 0.1))  # Mostrar ticks de 0.1 en 0.1 en el eje y
plt.show()
```

```python
# Curva interpolada (11 puntos)
# Niveles estándar
R = np.linspace(0,1,11) # Crea el array [0, 0.1, 0.2, ..., 1.0]
P_int = []
for r in R:
    # máximo precision para recall >= r
    values = [p for (rc,p) in zip(recall, precision) if rc >= r]
    P_int.append(max(values) if values else 0.0)
```

```python
# Plot interpolado
plt.figure(figsize=(8, 5))
plt.plot(R, P_int, 'o-')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Gráfico interpolado')
plt.legend()
plt.xticks(np.arange(0, 1.1, 0.1))  # Mostrar ticks de 0.1 en 0.1 en el eje y
plt.yticks(np.arange(0, 1.1, 0.1))  # Mostrar ticks de 0.1 en 0.1 en el eje y
plt.grid(True)
plt.show()
```
